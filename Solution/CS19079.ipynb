# Data Preprocessing

## 1) Dropping Irrelevant feature

```python
df.drop('Seat No.', axis = 1, inplace = True)
df.columns
```

```python
# print(f'General check for null values \n{df.isna().any()}\n')
print(f'Names of columns that contain null values: \n{df.columns[df.isna().any()].tolist()}\n')
print(f'Number of columns containing missing values: \n{len(df.columns[df.isna().any()].tolist())}\n')
print(f'Number of missing values in each column: \n{df.isna().sum()}')
    
```

### Removing rows containing null values of columns which contain less no.of null values

# Model 1

```python
from sklearn.model_selection import train_test_split

m1_df = df_1
print(m1_df.shape)

m1_X_train, m1_X_test, m1_y_train, m1_y_test = train_test_split(m1_df, y, test_size = 0.2)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr1 = LinearRegression()
lr1.fit(m1_X_train, m1_y_train)

lr1.score(m1_X_train, m1_y_train), lr1.score(m1_X_test, m1_y_test)
```

### KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor

knn1 = KNeighborsRegressor(n_neighbors= 10)
knn1.fit(m1_X_train, m1_y_train)

knn1.score(m1_X_train, m1_y_train), knn1.score(m1_X_test, m1_y_test)
```

### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

DT1 = DecisionTreeRegressor( max_depth=5 ) # deepening the DT causing more overfitting
DT1.fit(m1_X_train,m1_y_train)

DT1.score(m1_X_train, m1_y_train), DT1.score(m1_X_test, m1_y_test)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

RF1 = RandomForestRegressor(n_estimators=20, max_depth=10, criterion='mse')
RF1.fit(m1_X_train,m1_y_train)

RF1.score(m1_X_train, m1_y_train), RF1.score(m1_X_test, m1_y_test)
```
